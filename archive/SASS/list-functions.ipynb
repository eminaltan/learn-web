#### Sass List Functions

Kaynak: [W3 Schools](https://www.w3schools.com/sass/sass_functions_list.php)

SASS'da listeler immutable özelliğe sahiptir. Yani list fonksiyonları yeni bir liste döndürür. Orjinal liste içeriği değiştirilemez.


Liste öğelerinin index değeri 1 rakamından başlar. 0'dan başlamaz. (SASS string fonksiyonlarında olduğu gibi.)


Aşağıda SASS'da kullanılan list yöntemlerine ve örneklerine yer verilmiştir.

- `append(list, value, [separator])`: Listenin sonuna yeni bir liste değeri ekler. separator comma, space metinlerinden oluşabilir, varsayılan değeri `auto`'dur.

    **Ex.**
    `append((a b c),d)`; 
    
    Result: a b c d

    `append((a b c), (d), comma)`; /* `[separator]` comma ile ifade edilmiş. */

    Result: a,b,c,d


- `index(list, value)`: `value` değerinin `list` içerisindeki index pozisyon değerini verir. Eğer `value` liste içerisinde yoksa `null` olarak geri değer döndürülür.

    **Ex.**
    `index(a b c, b)`;

    Result: 2

    `index(a b c, f)`; /* f değeri liste içerisinde olmadığından `null` değeri geri döndürülüyor. */
    
    Result:null;


- `is-bracketed(list)`: `list` içeriğinin köşeli parantezden oluşup/oluşmadığını sorgular. Eğer köşeli parantezden oluşuyorsa sonuç `true` olarak döner aksi durumda `false`.

    **Ex.**
    `is-bracketed([a b c])`;

    Result: true

    `is-bracketed(a b c)`;
    
    Result: false


- `join(list1, list2, [separator, bracketed])`: `list2`'yi `list1`'in sonuna ekler. `[seperator]` comma veya space metinlerinden oluşabilir. Varsayılan değeri `auto`'dur. `bracketed` varsayılan olarak `auto`'dur. Fakat `true` ve `false` olarak ayarlanabilir. `true` olarak ayarlanması durumunda yeni liste köşeli parantez içerisine alınarak geri döndürülür.

    **Ex.**
    `join(a b c, d e f)`;

    Result: a b c d e f

    `join((a b c), (d e f), comma)`;

    Result: a,b,c,d,e,f

    `join(a b c, d e f, $bracketed: true)`;

    Result: [a b c d e f]


- `length(list)`: `list`'in uzunluğunu döndürür.

    **Ex.**
    `lis(5, 7 ,9, 0, -3, -7)`;

    Result: 6


- `list-separator(list)`: `list`'de kullanılan separator türünü döndürür.

    **Ex.**
    `list-separator(a b c)`;

    Result: "space"

    `list-separator(a,b,c)`;

    Result: "comma"


- `nth(list, n)`: `n` ile belirtilen index'e ait değeri `list` içerisinden döndürür.

    **Ex.**
    `nth(a b c, 3)`;

    Result: c


- `set-nth(list, n, value)`: `list` içerisinde `n` ile  belirtilen index'e ait değeri ayarlamaya yarar.

    **Ex.**
    `set-nth(a b c, 2, x)` /* Liste içerisindeki 2.elementin içeriğin x olarak ayarla. */

    Result: a x c


- `zip(lists)`: Liste içeriğini kombine ederek çıktı oluşturur.

    **Ex.**
    `zip(1px 2px 3px, solid dashed dotted, red green blue)`;

    Result: 1px solid red, 2px dashed green, 3px dotted blue
