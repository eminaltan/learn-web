#### Sass Numeric Functions

Kaynak: [W3 Schools](https://www.w3schools.com/sass/sass_functions_numeric.php)

SASS'da numeric değerleri manipule etmek için numeric fonksiyonlardan faydalanılır.

Aşağıda SASS'da kullanılan numeric yöntemlerine ve örneklerine yer verilmiştir.

- `abs(number)`: `number` değişenin mutlak değerini verir.

    **Ex.**
    `abs(15)`;
    
    Result: 15

    `abs(-15)`;

    Result: 15


- `cell(number)`: `number` değerini kendine en yakın **yuları** integer değerine yuvarlar.

    **Ex.**
    `cell(15.20)`;

    Result: 16


- `comparable(num1, num2)`: `num1` ve `num2` değerlerinin kıyaslanıp kıyaslanamyacağını hesaplar. `true` veya `false` değer döndürür.

    **Ex.**
    `comparable(15px, 10px)`;

    Result: true

    `comparable(20mm, 1cm)`;
    
    Result: true

    `comparable(35px, 2em)`;

    Result: false

- `floor(number)`: `number` değerini kendine en yakın **aşağı** integer değere yuvarlar.

    **Ex.**
    `floor(15.80)`;

    Result: 15


- `max(number...)`: `number` değişken listesindeki en büyük sayısal değeri döndürür.

    **Ex.**
    `max(5, 7 ,9, 0, -3, -7)`;

    Result: 9


- `min(number...)`: `number` değişken listesindeki en küçük sayısal değeri döndürür.

    **Ex.**
    `min(5, 7 ,9, 0, -3, -7)`;

    Result: -7


- `percentage(number)`: `number` değerini 100 ile çarparak percentage değere dönüştürür.

    **Ex.**
    `percentage(1.2)`;

    Result: 120


- `random()`: 0 ile 1 arasında random sayı döndürür.

    **Ex.**
    `random()`;

    Result: 0.45673


- `random(number)`: 1 ile `number` değeri arasında random sayı döndürür.

    **Ex.**
    `random(6)`;

    Result: 4

- `round(number)`: `number` değerini en yakın integer değerine döndürür.

    **Ex.**
    `round(15.20)`;
    
    Result: 15

    `round(15.80)`;

    Result: 16
