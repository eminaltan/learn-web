#### Sass String Functions

Kaynak: [W3 Schools](https://www.w3schools.com/sass/sass_functions_string.php)

⚠️ SASS'da String'lerindeki karakterlerin indexlemesi 0'dan değil 1'den başlar.

Aşağıda SASS'da kullanılan string yöntemlerine ve örneklerine yer verilmiştir.

- `quote(string)`: `string` değişkenine alıntılama işareti ekler.

    **Ex.**
    `quote(Hello world)`;
    
    Result: "Hello world"


- `str-index(string, substring)`: `string` içerisinde `substring`'i arar ve uyuşan ilk kısıma ait index değerini geri döndürür.

    **Ex.**
    `str-index("Hello world", "H")`;

    Result: 1


- `str-insert(string, insert, index)`: `string` içerisinde `insert` değerini ekler. Ekleme işlemini yeri `index` değeri ile belirlenir.

    **Ex.**
    `str-insert("Hello world!", "wonderful", 6)`;

    Result: "Hello wonderful world"


- `str-length(string)`: `string` değerinin uzunluğunu verir.

    **Ex.**
    `str-length("Hello world!")`;

    Result: 12


- `str-slice(string,start,end)`: `string` değerinin `start` ve `end` ile belirtilen aralıktaki değeri geri döndürür.

    **Ex.**
    `str-slice("Hello world!",2 ,5)`;

    Result: "ello"


- `to-lower-case(string)`: `string` değerini küçük harflere dönüştürerek geri döndürür.

    **Ex.**
    `to-lower-case("Hello World!")`;

    Result: "hello world!"


- `to-upper-case(string)`: `string` değerini büyük harflere dönüştürerek geri döndürür.

    **Ex.**
    `to-upper-case("Hello World1)`;

    Result: "HELLO WORLD"


- `unique-id()`: Random ***unique*** string üretir ve döndürür.

    **Ex.**
    `uniqure-id()`;

    Result: tyghefnsv


- `unquote(string)`: `string` değerinde varsa alıntılama işaretini kaldırır.

    **Ex.**
    `unquote("Hello world!")`;
    
    Result: Hello
