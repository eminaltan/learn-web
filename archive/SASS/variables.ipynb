#### SASS Variable Scope

SASS'da variable'lar local ve global olmak üzere ikiye ayrılır. 💡Aslında tüm proglama dilleri için geçerli bir durumdur bu.

N.Ş.A.'da global variable'lara her yerden erişim mümkün iken local variabler'lar bulunduğu scope içerisinde kullanılır. Scope dışında kullanılamaz.

Global variable'lar deklarasyonların veya ruler'rın daima dışında tanımlanırlar.

**Ex.**

```scss
$myColor: red;

h1 {
  /* $mycolor override ediliyor, yeni değeri scope dışında kullanılamaz. */
  $myColor: green;
  color: $myColor;
}

p {
  color: $myColor;
}
```

Yukarıdaki scss'in çıktısı aşağıdaki gibi olacaktır.

```css
h1 {
  color: green;
}

p {
  color: red;
}
```

💡Eğer local bir variable bulunduğu scope dışında kullanılmak istenirse variable'ın sonuna `!global` KW eklenir.

```scss
$myColor: red;

h1 {
  /* $mycolor override ediliyor, yeni değeri scope dışında kullanılabilir. Çünkü !global KW variable'ın global olduğunu scope dışında da kullanılabileceğini ifade ediyor. */
  $myColor: green !global;
  color: $myColor;
}

p {
  color: $myColor;
}
```

Yukarıdaki scss'in çıktısı aşağıdaki gibi olacaktır.

```css
h1 {
  color: green;
}

p {
  color: green;
}
```
